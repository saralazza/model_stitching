```python
# --- Notebook cell for ensembling ---
import pandas as pd
import numpy as np
import json

# 1. Load both submissions
try:
    df_mlp = pd.read_csv("/kaggle/input/submission-ensamble/submission (29) (1).csv")
    df_vae_mlp = pd.read_csv("/kaggle/input/submission-ensamble/submission (29) (1).csv")

except FileNotFoundError:
    print("Make sure both submission files are in the /kaggle/working/ directory")
    raise

# 2. Parse the JSON embeddings
embs_mlp = np.array([json.loads(e) for e in df_mlp['embedding']])
embs_vae_mlp = np.array([json.loads(e) for e in df_vae_mlp['embedding']])

# 3. Average the embeddings
# You can also try a weighted average, e.g., (0.7 * embs_mlp) + (0.3 * embs_vae_mlp)
avg_embs = (0.65 * embs_mlp) + (0.35 * embs_vae_mlp)

# 4. Create the new ensembled submission file
embedding_json_list = [json.dumps(embedding.tolist()) for embedding in avg_embs]
df_ensembled = pd.DataFrame({
    'id': df_mlp['id'],
    'embedding': embedding_json_list
})

df_ensembled.to_csv("submission_ensembled10.csv", index=False)

print("Ensembled submission created!")
print(df_ensembled.head())
```
